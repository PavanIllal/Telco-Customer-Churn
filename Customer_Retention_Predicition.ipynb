```python
cols = ["MultipleLines","InternetService","OnlineSecurity","OnlineBackup","DeviceProtection"
      ,"TechSupport","StreamingTV","StreamingMovies","Contract","PaperlessBilling","PaymentMethod"]
```

```python
for col in cols:
    print(df[col].value_counts())
```

```python
for col in cols:
    value_count = df[col].value_counts()
    proportion = df[col].value_counts(normalize=True)
    fig, axes = plt.subplots(1, 2, figsize=(10,5))

    # Bar plot (proportion)
    axes[0].bar(proportion.index, proportion.values)
    axes[0].set_title(f"Distribution of {col}")
    axes[0].set_xlabel(col)
    axes[0].set_ylabel("Proportion")

    if col == "PaymentMethod":
        axes[0].tick_params(axis='x', rotation=30)

    # Pie chart (percentage)
    axes[1].pie(
        value_count.values,
        labels=value_count.index,
        autopct="%1.1f%%"
    )
    axes[1].set_title(f"{col} Percentage")

    plt.tight_layout()
    plt.show()
```
